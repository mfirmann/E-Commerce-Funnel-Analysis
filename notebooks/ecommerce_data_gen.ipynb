```
import pandas as pd
import random
import datetime
import uuid

# Konfigurasi Jumlah Data
NUM_USERS = 5000  # Total user unik
START_DATE = datetime.date(2024, 1, 1)

# Persiapan List
devices = ['Android', 'iOS', 'Web']
sources = ['Instagram Ads', 'Google Ads', 'Organic', 'Direct']
events_flow = ['homepage_view', 'product_view', 'add_to_cart', 'checkout_start', 'purchase_success']

data = []

print("Sedang men-generate data transaksi... Mohon tunggu.")

for _ in range(NUM_USERS):
    # 1. Profil User
    user_id = str(uuid.uuid4())[:8] # ID user acak
    device = random.choices(devices, weights=[50, 30, 20])[0] # 50% user Android
    source = random.choice(sources)
    current_time = datetime.datetime.combine(START_DATE, datetime.time(random.randint(0, 23), random.randint(0, 59)))

    # 2. Simulasi Perjalanan User (Funnel Journey)
    # Setiap user PASTI masuk homepage
    has_churned = False # Apakah user kabur?

    for step, event_name in enumerate(events_flow):
        if has_churned:
            break # Kalau sudah kabur, stop record event selanjutnya

        # Catat event
        data.append({
            'user_id': user_id,
            'event_time': current_time,
            'event_type': event_name,
            'device': device,
            'source': source
        })

        # Tambah waktu sedikit untuk event berikutnya (biar logis)
        current_time += datetime.timedelta(seconds=random.randint(10, 300))

        # 3. LOGIKA "DROP-OFF" (Pola Rahasia)
        # Tentukan peluang user lanjut ke tahap berikutnya

        prob_next = 0.0

        if event_name == 'homepage_view':
            prob_next = 0.70 # 70% user lanjut lihat produk
        elif event_name == 'product_view':
            prob_next = 0.40 # 40% user lanjut add to cart
        elif event_name == 'add_to_cart':
            prob_next = 0.60 # 60% user lanjut checkout
        elif event_name == 'checkout_start':
            # === DISINI KITA TANAM BUG ANDROID ===
            if device == 'Android':
                prob_next = 0.10 # Cuma 10% user Android yang berhasil bayar (BUG PARAH!)
            else:
                prob_next = 0.90 # User iOS/Web lancar jaya (90% sukses)

        # Kocok dadu: Lanjut atau Kabur?
        if random.random() > prob_next:
            has_churned = True

# Buat DataFrame
df = pd.DataFrame(data)

# Simpan ke CSV
filename = "ecommerce_funnel.csv"
df.to_csv(filename, index=False)

print(f"✅ Selesai! {len(df)} baris data berhasil dibuat.")
print(f"File tersimpan dengan nama: {filename}")
print("Contoh 5 data teratas:")
print(df.head())
```
